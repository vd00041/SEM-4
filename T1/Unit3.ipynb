# Unit -3 - Feature creation & Feature Transformantion

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'A':[10,20,30,40,50,60,70,80],
    'B':[1,3,2,4,5,6,7,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
df
```

```python
df['C']=df['A']+df['B']
```

```python
df
```

```python
df['C']+4
```

```python
df['Result']= np.where(df['C']>50,'Pass','Fail')
df
```

```python
df["Grade"]=np.where(df['C']>80,'A',
                     np.where(df["C"]>60,'B',
                              np.where(df['C']>50,'C',
                                       np.where(df['C']>40,'D','E'))))
df
```
