````python
df.drop('loan_id',axis=1,inplace=True)
````

````python
df.isna().sum()
````

````python
df.drop_duplicates(inplace=True)
````

````python
# Calculate Q1, Q3, IQR for income_annum
Q1 = df['income_annum'].quantile(0.25)
Q3 = df['income_annum'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Drop outliers
df = df[(df['income_annum'] >= lower_bound) & (df['income_annum'] <= upper_bound)]
````

````python
df['total_assets_value']= df['residential_assets_value'] + df['commercial_assets_value']+df['bank_asset_value']+df['luxury_assets_value']
````

````python
df.drop(['residential_assets_value','commercial_assets_value','bank_asset_value','luxury_assets_value'],axis=1,inplace=True)
````

````python
df
````

````python
x = df.drop('loan_status',axis=1)
y = df['loan_status']
x = pd.get_dummies(x,drop_first=True)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
````

````python
#create knn(neighbors =10), decisiontress(criterion = entropy,random state=42),
#randomforst(criterion = entropy,random state=42),svc (c=20) 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt

knn_acc = []
dt_acc = []
rf_acc = []
svc_acc = []

test_sizes = [0.1, 0.2, 0.3, 0.4]

for test_size in test_sizes:
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
    
    # KNN
    knn = KNeighborsClassifier(n_neighbors=10)
    knn.fit(x_train, y_train)
    y_pred_knn = knn.predict(x_test)
    knn_acc.append(accuracy_score(y_test, y_pred_knn))
    
    # Decision Tree
    dt = DecisionTreeClassifier(criterion='entropy', random_state=42)
    dt.fit(x_train, y_train)
    y_pred_dt = dt.predict(x_test)
    dt_acc.append(accuracy_score(y_test, y_pred_dt))
    
    # Random Forest
    rf = RandomForestClassifier(criterion='entropy', random_state=42)
    rf.fit(x_train, y_train)
    y_pred_rf = rf.predict(x_test)
    rf_acc.append(accuracy_score(y_test, y_pred_rf))
    
    # SVC
    svc_model = SVC(C=20)
    svc_model.fit(x_train, y_train)
    y_pred_svc = svc_model.predict(x_test)
    svc_acc.append(accuracy_score(y_test, y_pred_svc))

# Plot
plt.plot(test_sizes, knn_acc, label='KNN')
plt.plot(test_sizes, dt_acc, label='Decision Tree')
plt.plot(test_sizes, rf_acc, label='Random Forest')
plt.plot(test_sizes, svc_acc, label='SVC')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.title('Model Accuracy vs Test Size')
plt.legend()
plt.show()
    
    

````
